# Равномерное распределение 

```python
%matplotlib inline
```

```python
from scipy.stats import uniform
from matplotlib import pyplot as plt
import numpy as np
```

```python
a = -1
b = 1
x = np.linspace(a - 1, b + 1, 100)
```

```python
dist = uniform(loc=a, scale=abs(a - b))
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Uniform Distribution function')
plt.plot(x, dist.cdf(x), color='r', label=r'F({0}, {1})'.format(a, b))
plt.legend()
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Probability density function')
plt.plot(x, dist.pdf(x), color='b', label=r'f({0}, {1})'.format(a, b))
plt.legend()
```

# Пуассоновское распределение 

```python
from scipy.stats import poisson
```

```python
mu = 10
x = np.arange(-1, 20)
```

```python
dist = poisson(mu)
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Poisson Distribution function')
plt.plot(x, dist.cdf(x), color='b', label=r'F({0})'.format(mu))
plt.legend()
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Probability density function')
plt.plot(x, dist.pmf(x), color='r', label=r'f({0})'.format(mu))
plt.legend()
```
